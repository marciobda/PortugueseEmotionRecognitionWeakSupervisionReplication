```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
df_test = pd.read_csv('./test.csv', sep='\t')
print(df_test.head())
```

```python
df_test = df_test.dropna(subset=['categoria'])
print(df_test.head())
```

```python
df_test['categoria'] = df_test['categoria'].apply(ast.literal_eval)
df_test['categoria'].dtypes
```

```python
df_exploded = df_test.explode('categoria')
value_counts = df_exploded['categoria'].value_counts()
value_counts
```

```python
value_counts.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Categoria')
plt.ylabel('Count')
plt.title('Count of Lines by Categoria')
plt.show()
```
